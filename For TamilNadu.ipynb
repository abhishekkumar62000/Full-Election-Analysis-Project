```python
#WORKING CODE - for TN 

import pandas as pd

#For 1989-1999 datasets

files=['1989.csv','1991.csv','1996.csv','1998.csv','1999.csv']
year_parties=[]

for file in files:
    df=pd.read_csv(file)

    y1=df[df['TERRITORY']=='TAMIL NADU']
    y1=y1[['No.', '%', 'CONSTITUENCY', 'PARTY']]

    j=-1
    unique=list(y1.CONSTITUENCY.unique())
    check=[]
    votes_percent=[]
    party=[]

    for i in y1.index:
        const=y1['CONSTITUENCY'][i]
        if(const in unique and const not in check):
            j+=1
            check.append(const)
            votes_percent.insert(j, y1['%'][i])
            party.insert(j, y1['PARTY'][i])

        if(const in unique and const in check):
            if(y1['%'][i]>votes_percent[j]):
                votes_percent[j]=y1['%'][i]
                party[j]=y1['PARTY'][i]

    f1=pd.DataFrame({'Constituency': check, file.split('.')[0]: party})
    year_parties.append(party)
    #print(f1)
    
#For 2004 dataset

df=pd.read_csv('2004.csv')

y1=df[df['TERRITORY']=='TAMIL NADU']
y1=y1[['No', '% OF TOTAL VOTES', 'CONSTITUENCY', 'PARTY']]

j=-1
unique=list(y1.CONSTITUENCY.unique())
check=[]
votes_percent=[]
party=[]

for i in y1.index:
    const=y1['CONSTITUENCY'][i]
    if(const in unique and const not in check):
        j+=1
        check.append(const)
        votes_percent.insert(j, y1['% OF TOTAL VOTES'][i])
        party.insert(j, y1['PARTY'][i])
        
    if(const in unique and const in check):
        if(y1['% OF TOTAL VOTES'][i]>votes_percent[j]):
            votes_percent[j]=y1['%'][i]
            party[j]=y1['PARTY'][i]

f2=pd.DataFrame({'Constituency': check, '2004': party})
#print(f2)

#2009

cols=['STATE','CONSTITUENCY','CANDIDATE','PARTY','% OF TOTAL VOTES']
df = pd.read_csv("2009.csv",usecols=cols)
df=df[df['STATE']=='Tamil Nadu']

ConstituenciesList = set(df['CONSTITUENCY'])
Candidates_constituency = {}
Winner={}

for item in ConstituenciesList:
    if item not in Candidates_constituency:
        Candidates_constituency[item]=df[df['CONSTITUENCY']==item]
    
for dataframe in Candidates_constituency:
    Candidates_constituency[dataframe]=Candidates_constituency[dataframe].sort_values(by=['% OF TOTAL VOTES'],ascending=False)
    if dataframe not in Winner:
        df_temp=Candidates_constituency[dataframe].head(1)
        WinnerParty=str(df_temp['PARTY'])
        WinnerParty=WinnerParty.split(' ')[4][:-6]
        Winner[dataframe]=WinnerParty

Const=[]
Party2009=[]
for item in Winner:
    Const.append(item)
    Party2009.append(Winner[item])
    
f3=pd.DataFrame({'Constituency': Const, '2009': Party2009})
#print(f3)

#2014

cols=['TERRITORY','CONSTITUENCY','CANDIDATE','PARTY','% OF TOTAL POLLED']
df = pd.read_csv("2014.csv",usecols=cols)
df=df[df['TERRITORY']=='Tamil Nadu']

ConstituenciesList = set(df['CONSTITUENCY'])
Candidates_constituency = {}
Winner={}

for item in ConstituenciesList:
    if item not in Candidates_constituency:
        Candidates_constituency[item]=df[df['CONSTITUENCY']==item]
    
for dataframe in Candidates_constituency:
    Candidates_constituency[dataframe]=Candidates_constituency[dataframe].sort_values(by=['% OF TOTAL POLLED'],ascending=False)
    if dataframe not in Winner:
        df_temp=Candidates_constituency[dataframe].head(1)
        WinnerParty=str(df_temp['PARTY'])
        WinnerParty=WinnerParty.split(' ')[4][:-6]
        Winner[dataframe]=WinnerParty

Const=[]
Party2014=[]
for item in Winner:
    Const.append(item)
    Party2014.append(Winner[item])
    
f4=pd.DataFrame({'Constituency': Const, '2014': Party2014})
#print(f4)


FinalDF1 = pd.DataFrame ({'Constituency':check,'1989':year_parties[0],'1991':year_parties[1],'1996':year_parties[2],'1998':year_parties[3],'1999':year_parties[4],'2004':party})
print (FinalDF1)

FinalDF2 = pd.DataFrame ({'Constituency':Const,'2009':Party2009,'2014':Party2014})
print (FinalDF2)
```
